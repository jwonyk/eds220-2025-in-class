# Geopandas

GeoPandas is a Python library that extends the `pandas` by adding support for geospatial data.

## Reproducible file paths

```python
import os
import geopandas as gpd
import matplotlib.pyplot as plt
```

To import data we first use `os` package to create a reproducible file path:

```python
fp = os.path.join('data', 'gbif_sus_scrofa_california', 'gbif_sus_scrofa_california.shp')
fp
```

```python
pigs = gpd.read_file(fp)
pigs.head()
```

```python
# Import states data
fp = os.path.join('data','ca_state_boundary','ca_state_boundary.shp')
ca_boundary = gpd.read_file(fp)
ca_boundary.head()
```

## `GeoSeries` and `GeoDataFrame`
The **geometry column** holds the geometry (point, polygon, etc.) of each spatial feature. This geometry column is of type `geopandas.GeoSeries`. Columns in the `geopandas.GeoDataFrame` with attributes about the features are `pandas.Series` like in a regular `pandas.DataFrame`.

```python
# Check the data type of the pigs dataframe 
print(type(pigs))

# Check the data type of the geometry column
print(type(pigs.geometry))

# Check the data type of the gbifID column
print(type(pigs.gbifID))
```

```python
# Check the data type of each column
pigs.dtypes
```

```python
# Check the data type of geometry column
pigs.geom_type
```

What is the geometry type of the single feature in the CA state boundary?

```python
ca_boundary.geom_type
```

## CRS

**Coordinate Reference System (CRS)**: the instructions to locate each spatial feature of our data frame on the surface of the Earth. We access the CRS of a geopandas.GeoDataFrame using the crs attribute.

```python
# Access the CRS of the GeoDataFrame
pigs.crs
```

```python
# Examine CRS details
print('Ellipsoid:', pigs.crs.ellipsoid)
print('Datum:', pigs.crs.datum)
print('Is geographic?:', pigs.crs.is_geographic)
print('Is projected?:', pigs.crs.is_projected)
```

```python
# Obtain the geographic extent of the geo-dataframe
pigs.total_bounds
```

## Extent
The **extent** of the geo-dataframe is the bounding box covering all the spatial features in our geo-dataframe. This is formed by finding the points that are furthest west, east, south, and north.

```python
# Examine CRS and extent of CA boundary
print('CRS: ', ca_boundary.crs, '\n')
print(ca_boundary.total_bounds) 
```

## Data wrangling
GeoPandas is conveniently built on top of `pandas`, so we may use everything we have learned about data selection, wrangling, and modification for a pandas.DataFrame to wrange geopandas.GeoDataFrames.



# Examine pig observations 

```python
pigs['year'].value_counts().sort_index()
```

```python
# Select data from 2020 onwards
pigs_recent = pigs[pigs['year'] >= 2020]

# Check length of original dataframe
print('Total number of observations: ', len(pigs))

# Check length of new dataframe
print('Number of observations since 2020: ', len(pigs_recent))
```

## `matplotlib`

We can interact with `matplotlib` via its `pyplot` interface.

Matplotlib graphs the data in **figures** that can have one or more **axes**. The axes is *only* the area specified by the x axis and the y axis what is plotted in it. A figire can include multiple axes in it.

Tp create a blank figure:

```python
# Initialize empty figure (fig) and axis (ax)
fig, ax = plt.subplots() # Return two objects: fig and ax

# Display figure
plt.show()
```

## Adding a layer

When using `matplotlib` it can be useful to think of creating a plot as adding layers on an axis.

Example:

The first layer we want to add to our axis is the `pigs_recent` point data

```python
# Initialize empty figure and axis
fig, ax = plt.subplots()

# Add pigs point 
pigs_recent.plot(ax = ax)

# Display figure
plt.show()
```

## Customization

Some customization can be done directly in `plot()` method for the dataframe. Some other is done by updating attributes of the axis = 

```python
# Initialize empty figure and axis
fig, ax = plt.subplots()

# Add data to axis
pigs_recent.plot(ax = ax, # Add plot to axis
                 alpha = 0.5, # Adjust transparency 
                 color = 'brown') # Update color

# Update axis
ax.set_title('Reported "Sus scrofa" sightings in CA (2020 - 2023)')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Display figure
plt.show()
```

```python
# Initialize empty figure and axis
fig, ax = plt.subplots()

# Add California Boundary
ca_boundary.plot(ax = ax,
                 color = 'none',
                 edgecolor = 'black')

# Add data to axis
pigs_recent.plot(ax = ax, # Add plot to axis
                 alpha = 0.5, # Adjust transparency 
                 color = 'brown', # Update color
                 edgecolor = 'yellow') 

# Update axis
ax.set_title('Reported "Sus Scrofa" sightings in CA (2020 - 2023)')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Display figure
plt.show()
```
