```python
import numpy as np
import pandas as pd

# Load Palmer penguins data
URL = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/main/inst/extdata/penguins.csv'
penguins = pd.read_csv(URL)

penguins.head()
```

## Check-in
1. Use the max() method for pandas.DataFrames to calculate the maximum value of a penguin’s body mass by year and species.
2. Use (1) to display the highest body masses per year and species as a bar plot in descending order.

```python
# Use the max() method for pandas.DataFrames to calculate the maximum value of a penguin’s body mass by year and species.
(penguins.groupby(['year', 'species'])
         .body_mass_g
         .max())
```

```python
# Use (1) to display the highest body masses per year and species as a bar plot in descending order.
(penguins.groupby(['year', 'species'])
         .body_mass_g
         .max()
         .sort_values()
         .plot(kind = 'barh',
               title = 'Maximum Penguin Body Mass per Species per Year',
               ylabel = ('Year, Species'),
               xlabel = ('Body Mass (g)')))
```
